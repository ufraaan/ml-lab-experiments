```python
# part a expected value

import numpy as np

x = np.array([10, 20, 30])
f = np.array([2, 3, 5])

expected_value = np.sum(x * f) / np.sum(f)

print("Expected Value:", expected_value)
```

```python
# part b mean

import numpy as np

arr = np.array([1, 2, 3, 4, 5])

mean_value = np.mean(arr)

print("Mean:", mean_value)
```

```python
# part c standard deviation

import numpy as np

arr = np.array([1, 2, 3, 4, 5])

std_dev = np.std(arr)

print("Standard Deviation:", std_dev)
```

```python
# part d variance

import numpy as np

arr = np.array([1, 2, 3, 4, 5])

variance = np.var(arr)

print("Variance:", variance)
```

```python
# part e covariance

import numpy as np

arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([5, 4, 3, 2, 1])

covariance = np.cov(arr1, arr2)[0, 1]

print("Covariance:", covariance)
```

```python
# part f covariance matrix

import numpy as np

psychology = np.array([80, 63, 100])
history = np.array([70, 20, 50])

cov_matrix = np.cov(psychology, history)

print("Covariance Matrix:")
print(cov_matrix)
```
