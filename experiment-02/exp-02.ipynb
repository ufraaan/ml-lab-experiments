```python
"""
exp2

a. expected value
input format:


| value (x) | frequency (f) |
| 10 | 2 |
| 20 | 3 |
| 30 | 5 |

ALGORITHM:
1. start the program
2. initialise a list of values
3. intiialise a list of corresponding frequencies
4. calculate the total frequcny by summing all freqency values.
5. intiailise a variable to strore the sum of X x f
6. for each value and its frequency:
    - multiply the vlaue by its frequency
    - add the result to the sum
7. divide the total of X x f by the total frequency
8. display the expected value
9. stop the program

EXPECTED OUTPUT:
expected value: 23.0
"""

# step 1: initialise step of values
values = [10, 20, 30]

# step 2: initialise the corresponding frequencies
freq = [2, 3, 5]


# step 3: store sum of all freqs
total_freq = sum(freq)


# step 4: intialise var to store sum of x*f
total_xf = 0


# step 5: store for each x * f in total_xf
for i in range (len(values)):
    total_xf += values[i] * freq[i]

# step 6: 
final = total_xf / total_freq


# step 7: 
print(f"Expected Value is: {final}")
```

```python
"""
exp2

b. mean
input format:

[1, 2, 3, 4, 5]

ALGORITHM:
1. Start the program
2. Input the array
3. Convert to NumPy array
4. Compute mean using np.mea(array)
5. print the result
6. end the program

EXPECTED OUTPUT:
mean value: 3.0
"""


# step 1: initialise array
arr = [1, 2, 3, 4, 5]

# step 2: convert to numpy array
import numpy as np
npArr = np.array(arr)

# step 3: compute mean using np.mean()
meanValue = np.std(npArr)

# step 4: print the result
print(f"Mean value is: {meanValue}")
```
